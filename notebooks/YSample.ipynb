```python
import yfinance as yf

msft = yf.Ticker("MSFT")

# get stock info
msft.info

# get historical market data
hist = msft.history(period="max")

hist

```

```python
# minimum processing to be compatible with OHLCFigure
# hist['datetime'] = pandas.to_datetime(dataframe.timestamp, unit="s")

from bokeh.models import ColumnDataSource
from bokeh.io import show
from bokeh.plotting import output_notebook

output_notebook()

hist.columns = [c.lower() for c in hist.columns]
hist['datetime'] = hist.index

cds = ColumnDataSource(hist, name="OHLC")

from bokeh_ta.ohlc import ohlc
plot = ohlc(source=cds)


show(plot(plot_height=320, tools='pan, wheel_zoom', toolbar_location="left",
               x_axis_type="datetime", y_axis_location="right",
               sizing_mode="scale_width"))
```

```python
from pandas_ta import macd

macd_df = hist.ta.macd(fast=3, slow=6, signal=9)
macd_df['datetime'] = macd_df.index

cds = ColumnDataSource(macd_df, name="OHLC")

from bokeh_ta.macd import macd as macdplot
plot = macdplot(source=cds, fast=3, slow=6, signal=9)


show(plot(plot_height=320, tools='pan, wheel_zoom', toolbar_location="left",
               x_axis_type="datetime", y_axis_location="right",
               sizing_mode="scale_width"))
```
